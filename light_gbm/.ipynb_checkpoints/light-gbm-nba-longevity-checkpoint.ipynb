# Experimenting with LightGBM

Data: NBA Rookie Longevity<br>
Source: https://data.world/exercises/logistic-regression-exercise-1<br>
Question: Will each rookie last five years in the league?

```python
import lightgbm
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.metrics import confusion_matrix, accuracy_score, roc_auc_score, roc_curve
from sklearn.model_selection import train_test_split
import time
import xgboost
```

```python
sns.set(font_scale=1.2)
sns.set_style("whitegrid")
```

#### Load Data

```python
base_path = "/Users/aaronjones/Documents/github/classification-exercises/light_gbm/"
df = pd.read_csv(base_path+"nba_logreg.csv", header=0)
```

```python
print(df.shape)
print(df.columns)
```

```python
df = (
    df
    .rename(columns={"FG%": "FGpct", "3P Made": "3Pmade", "3P%": "3Ppct", "TARGET_5Yrs": "y"})
    .drop(columns="Name")
    .dropna()
)

print(df.shape)
print(df.columns)
```

#### Data Prep

```python
# split into features (X) and target (y)
X = df.drop(columns="y").to_numpy()
y = df["y"].values
```

```python
# split into train and test for both X and y
Xtrain, Xvalid, ytrain, yvalid = train_test_split(
    X, y, test_size=0.3, random_state=0, shuffle=True, stratify=y
)
```

```python
# percentage of classes for the whole, train, and test target vectors
target_vecs = {"all": df["y"].values, "ytrain": ytrain, "ytest": yvalid}
for k in target_vecs.keys():
    tmp = target_vecs[k]
    unique, counts = np.unique(tmp, return_counts=True)
    percents = np.array([round(c / len(tmp), 4) for c in counts])
    output = np.asarray((unique, percents)).T
    print("Name: "+ k)
    print(output)
```

#### LightGBM (Tag: 1)

```python
# put data in lightgbm format
dtrain1 = lightgbm.Dataset(Xtrain, label=ytrain)
dvalid1 = lightgbm.Dataset(Xvalid, label=yvalid, reference=dtrain1)
```

```python
# define:
# - basic parameters dictionary
# - an empty dictionary to house evaluation results
param1 = {
    'objective': 'binary', 
    'metric': ['auc', 'binary_logloss', 'binary_error'],
    'learning_rate': 0.01,
    'max_depth': 3,
    'num_leaves': 5
}

evals_result1 = {}
```

```python
# train light gbm model
stime1 = time.time()

lgb = lightgbm.train(
    param1,
    dtrain1,
    num_boost_round=150,
    valid_sets=[dtrain1, dvalid1],
    feature_name=list(df.columns[:-1]), 
    evals_result=evals_result1,
    verbose_eval=10
)

etime1 = time.time()
print("LightGBM train time is {} seconds.".format(round(etime1 - stime1, 2)))
```

```python
# plot metrics by tree iteration
# note: boosted models feature consecutively, not concurrently, built trees 
fig, axes = plt.subplots(nrows=3, ncols=1, figsize=(10, 5))

for idx, score in enumerate(param1["metric"]):
    
    # isolate list of scores
    tmp_train = evals_result1["training"][score]
    tmp_valid = evals_result1["valid_1"][score]
    # plot scores
    axes[idx].plot(tmp_train, c="blue", lw=2, label="Training")
    axes[idx].plot(tmp_valid, c="red", lw=2, label="Validation")
    # plot horizontal line of list score value
    # help visually identify non-improvement
    axes[idx].axhline(tmp_train[-1], c="blue", lw=1.5, ls="--", alpha=0.65)
    axes[idx].axhline(tmp_valid[-1], c="red", lw=1.5, ls="--", alpha=0.65)
    # add y-axis label
    axes[idx].set_ylabel(score)
    
    if idx != 2:
        
        # remove tick marks for all, but last subplot
        axes[idx].xaxis.set_major_formatter(plt.NullFormatter())

plt.suptitle("Metric by Tree Iteration")
plt.legend()
plt.show()
```

```python
# predict class probability for both training and validation data
ytrain_prob1 = lgb.predict(Xtrain)
yvalid_prob1 = lgb.predict(Xvalid)
```

```python
# determine the probability threshold for the best classification results
thresholds = np.linspace(0.05, 0.95, 19)
logistic_arrays1 = []
logistic_scores1 = []

for t in thresholds:
    tmp_class = [1. if p >= t else 0. for p in ytrain_prob1]
    logistic_arrays1.append(tmp_class) 
    logistic_scores1.append(roc_auc_score(ytrain, tmp_class))
    
index1 = np.argmax(logistic_scores1)
optimal_threshold1 = round(thresholds[index1], 2)

print("Probability Threshold: " + str(optimal_threshold1))
```

```python
# define predicted classes
# print evaluation metrics

ytrain_class1 = logistic_arrays1[index1]
yvalid_class1 = [1. if p >= optimal_threshold1 else 0. for p in yvalid_prob1]

print("""
------Training Metrics------\n
Confusion Matrix:\n{cntrain}
Accuracy:{acctrain}
ROC AUC Score:{auctrain}\n
------Validation Metrics------\n
Confusion Matrix:\n{cnvalid}
Accuracy:{accvalid}
ROC AUC Score:{aucvalid}""".format(
    cntrain=confusion_matrix(ytrain, ytrain_class1),
    acctrain=round(accuracy_score(ytrain, ytrain_class1), 3),
    auctrain=round(roc_auc_score(ytrain, ytrain_class1), 3),
    cnvalid=confusion_matrix(yvalid, yvalid_class1),
    accvalid=round(accuracy_score(yvalid, yvalid_class1), 3),
    aucvalid=round(roc_auc_score(yvalid, yvalid_class1), 3)
))
```

```python
# plot roc curve

fpr1, tpr1, thresholds1 = roc_curve(yvalid, yvalid_prob1, pos_label=1.)
auc_score1 = round(roc_auc_score(yvalid, yvalid_class1), 3)

plt.plot(fpr1, tpr1, label="ROC Curve (Area = %0.3f)" % auc_score1)
plt.plot([0, 1], [0, 1], "k--")
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.0])
plt.xlabel("FPR")
plt.ylabel("TPR")
plt.title("ROC Curve: LightGBM")
plt.legend(loc="lower right")
```
